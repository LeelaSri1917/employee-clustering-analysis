###**Regex-Based Cleaning of Company Names**
---

```python
df1 = df.copy()
```

- **Creates a separate copy of the Original Dataset**
- **Ensures the original data remains unchanged**

```python
import re
```

- **Import the Regular Expression (regex) library**
- **Regex helps in pattern-based text cleaning**

```python
def clean_string(s):
    if not isinstance(s, str):
        s = str(s)
    return re.sub('[^A-Za-z0-9 ]+', '', s)
```

- **Defines a Function to clean text values**
- **Converts non-string values (like NaN) into strings**
- **Prevents errors while applying regex**
- **Removes all characters except:**
  - **Uppercase letters (A–Z)**
  - **Lowercase letters (a–z)**
  - **Numbers (0–9)**
  - **Spaces**
- **Cleans tabs, symbols, punctuation, etc.**

```python
df1['company_hash'] = df1['company_hash'].apply(clean_string)
```

- **Applies the cleaning function to every value in the company_hash column**
- **Ensures consistent and clean company names**

###**Outlier Treatment Using 99th Percentile Capping Method**
---

#Why We Used Capping Instead of Removing Outliers
- **Salary and experience naturally have some High values.**
- **Completely removing them may:**
  - **Reduce dataset size**
  - **Remove valid High-Performing Employees**
- **Capping:**
  - **keeps all data points**
  - **Reduces Impcat of extreme values**
  - **Makes Distribution More stable**
  - **Improves Model Performance**
- **So we used 99Th Percentile capping**

---
###**Calculate 99th Percentile for CTC**
---

```python
ctc_upper_bound = df2['ctc'].quantile(0.99)
```

- **Finds the value below which 99% of salaries fall**
- **Top 1% Extreme salaries are considered Outliers**


---
###**Apply Clipping (Capping) on CTC**
---

```python
df2['ctc_capped'] = np.clip(df2['ctc'], df2['ctc'].min(), ctc_upper_bound)
```

>**What this Does:**
- **If Salary > 99th percentile : replace with upper Bound**
- **If Salary is Normal : Keep it unchanged**
- **Minimum value remains same**

---
###**Repeat Same for YOE**
---

```python
yoe_upper_bound = df2['YOE'].quantile(0.99)
df2['YOE_capped'] = np.clip(df2['YOE'], df2['YOE'].min(), yoe_upper_bound)
```

- **Caps Extreme Experience values**
- **Prevents unrealistic values like 120+ years**

##**Manual Clustering: Creating Designation, Class & Tier Level Salary Insights**
---
###**Why we are Doing This**
---
>**We are creating manual clustering groups based on different levels og aggregation**
---
>**1.DESIGNATION LEVEL**
- **Grouping by:**
  - **Company**
  - **Job Position**
  - **Years of Experience**
- **This gives very detailed salary insights**
---
>**CLASS LEVEL**
- **Grouping by:**
  - **Company**
  - **Job Position**
- **This gives mid-level salary trends (ignores experience variation)**
---
>**TIER LEVEL**
- **Grouping by:**

>**PURPOSE OF THIS STEP**

- **To Support manual Clustering logic**

```python
# Step 1: Group by company, job position, and years of experience (Designation Level)
grouped_summary_designation = df3.groupby(
    ['company_hash', 'job_position', 'YOE_capped']
)['ctc_capped'].agg(['mean', 'median', 'max', 'min', 'count']).reset_index()

grouped_summary_designation.rename(
    columns={'mean': 'mean_designation'},
    inplace=True
)


# Step 2: Group by company and job position (Class Level)
grouped_summary_class = df3.groupby(
    ['company_hash', 'job_position']
)['ctc_capped'].agg(['mean', 'median', 'max', 'min', 'count']).reset_index()

grouped_summary_class.rename(
    columns={'mean': 'mean_class'},
    inplace=True
)


# Step 3: Group by company only (Tier Level)
grouped_summary_tier = df3.groupby(
    ['company_hash']
)['ctc_capped'].agg(['mean', 'median', 'max', 'min', 'count']).reset_index()

grouped_summary_tier.rename(
    columns={'mean': 'mean_tier'},
    inplace=True
)
```

##**Merging Aggregated Salary Insights Back to Main Dataset**
---
###**Why we are Doing This**
---
>**After calculating salary summaries at:**
- **Designation Level (Company + Job + Experience)**
- **Class Level (Company + Job)**
- **Tier Level (Company Only)**
- **We Now merge the mean salary values back into the main dataset (df3)**
---

---
- **To Attach salary benchmarks to each employee record**
- **To Compare Individual Salary with:**

```python
# Step 2: Merge only the mean values with the original dataset

# Merge Designation Level Mean
df3 = df3.merge(
    grouped_summary_designation[['company_hash', 'job_position', 'YOE_capped', 'mean_designation']],
    on=['company_hash', 'job_position', 'YOE_capped']
)

# Merge Class Level Mean
df3 = df3.merge(
    grouped_summary_class[['company_hash', 'job_position', 'mean_class']],
    on=['company_hash', 'job_position']
)

# Merge Tier Level Mean
df3 = df3.merge(
    grouped_summary_tier[['company_hash', 'mean_tier']],
    on=['company_hash']
)
```

##**Creating Salary Comparison Flags (Designation, Class & Tier Levels)**
---
###**Why We are Doing This**
---
>**Now that we have salary benchmark averages:**
- **mean_designation**
- **mean_class**
- **mean_tier**
- **we create flags to compare each employee's salary (ctc_capped) with these benchmarks**
- **This Helps classify employees as:**
  - **Above average**
  - **Equal to Average**
  - **Below average**
---
>**Flag Logic**
| Condition     | Flag Value | Meaning          |
| ------------- | ---------- | ---------------- |
| Salary > Mean | 3          | Above Average    |
| Salary = Mean | 2          | Equal to Average |
| Salary < Mean | 1          | Below Average    |
---


```python
# Step 3: Create flags for Designation, Class, and Tier based on mean values

def designation_flag(row):
    if row['ctc_capped'] > row['mean_designation']:
        return 3
    elif row['ctc_capped'] == row['mean_designation']:
        return 2
    else:
        return 1


def class_flag(row):
    if row['ctc_capped'] > row['mean_class']:
        return 3
    elif row['ctc_capped'] == row['mean_class']:
        return 2
    else:
        return 1


def tier_flag(row):
    if row['ctc_capped'] > row['mean_tier']:
        return 3
    elif row['ctc_capped'] == row['mean_tier']:
        return 2
    else:
        return 1
```

>**You Now need to apply these functions to create actual flag columns:**

```python
df3['designation_flag'] = df3.apply(designation_flag, axis=1)
df3['class_flag'] = df3.apply(class_flag, axis=1)
df3['tier_flag'] = df3.apply(tier_flag, axis=1)
```

```python
from sklearn.neighbors import NearestNeighbors
```

- **Tests whether your dataset has natural cluster structure**
- **Compares Distance of real points vs randomly Generated points**
- **Returns a score between 0 and 1**
---
###**How to Interpret the Result**
---
| Hopkins Score | Meaning                                        |
| ------------- | ---------------------------------------------- |
| **≈ 0.5**     | Data is random → Not suitable for clustering ❌ |
| **0.6 – 0.7** | Weak clustering tendency                       |
| **0.7 – 0.8** | Good clustering tendency ✅                     |
| **> 0.8**     | Strong clustering structure 🔥                 |
| **≈ 1**       | Highly clusterable                             |
---
>**Before applying:**
- **K-Means**
- **Hierarchical Clustering**
- **The Dataset actually Contains Cluster structure**
- **Otherwise Clustering Results would be Meaningless**
---

```python
def hopkins_statistic(X):
    X = np.array(X)  # Ensure X is a numpy array
    d = X.shape[1]  # Number of dimensions
    n = len(X)  # Number of data points
    m = int(0.1 * n)  # Subset size (10% of the data points)

    nbrs = NearestNeighbors(n_neighbors=1).fit(X)
    rand_X = np.random.random((m, d)) * np.amax(X, axis=0)
    u_distances, _ = nbrs.kneighbors(rand_X, 2, return_distance=True)

    w_distances, _ = nbrs.kneighbors(X[np.random.choice(n, m, replace=False)], 2, return_distance=True)

    u_distances = u_distances[:, 1]
    w_distances = w_distances[:, 1]

    H = (np.sum(u_distances) / (np.sum(u_distances) + np.sum(w_distances)))
    return H

hopkins_score = hopkins_statistic(df_scaled)
print(f"Hopkins Statistic: {hopkins_score}")
```

#Observations:
- **Hopkins Statistic = 0.983**
- **This value is very close to 1.**
- **It means the data has strong clustering tendency.**
- **In simple terms, the data forms natural groups (clusters) rather than being random.**
- **Conclusion: The dataset is good for clustering analysis (like K-Means).**

###**Elbow Method – Selecting Optimal Number of Clusters**
---
- **Measures how tightly data points are grouped within a cluster**
- **Lower Inertia : More Compact Clusters**
- **Goal: Find the elbow point where reduction in inertia slows down**
---

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')

from yellowbrick.cluster import KElbowVisualizer

# Initialize the KMeans model
model = KMeans(random_state=42)

# Initialize the visualizer
visualizer = KElbowVisualizer(model, k=(1,10), metric='distortion')

# Fit the data to the visualizer
visualizer.fit(df_scaled)

# Finalize and show the figure
visualizer.show()
```

#Observations:
- **Best number of clusters : The best value of k is 5.**
- **Why k = 5? : At k = 5, the graph forms an "elbow", meaning adding more clusters after this does not improve results much.**
- **Distortion score : The score drops quickly from k = 1 to k = 5, then decreases slowly.**
- **Fit time:The time to compute clusters changes slightly, but becomes stable around k = 5.**
- **Conclusion: Using 5 clusters is a good choice for this dataset.**

###**Between-Cluster Sum of Squares (BCSS)**
---

```python
# Assuming df_scaled is your scaled dataframe
df_scaled_copy = df_scaled.copy()

# Adding cluster labels to the DataFrame
df_scaled_copy['kmeans_cluster'] = kmeans.labels_

# Between-Cluster Sum of Squares (BCSS)
def calculate_bcss(df, kmeans):
    cluster_centers = kmeans.cluster_centers_
    overall_mean = df.drop(columns='kmeans_cluster').mean(axis=0)
    bcss = 0
    for i, center in enumerate(cluster_centers):
        size = len(df[df['kmeans_cluster'] == i])
        bcss += size * np.sum((center - overall_mean) ** 2)
    return bcss

bcss = calculate_bcss(df_scaled_copy, kmeans)
print(f'Between-Cluster Sum of Squares (BCSS): {bcss}')
```

#Observations:
- **BCSS = 405,340.06**
- **BCSS measures how far the clusters are from each other.**
- **Higher BCSS means clusters are well separated.**
- **It shows how different the clusters are from one another.**
- **Your clusters are fairly well separated from each other.**

#Interpretation:
- **Higher BCSS : Better Cluster separation**
- **Lower BCSS : Centroids are close to overall Mean (Poor Separation)**
- **BCSS Capture How far clusters are from each other**
---
###**Relationship Between WCSS and BCSS**
---
- **Total Sum of Squares (TSS):**
  - **TSS = WCSS + BCSS**
- **Where:**
  - **WCSS - Compactness within Clusters**
  - **BCSS - Separation Between Clusters**
- **Good Clustering should have:**
  - **Low WCSS**
  - **High BCSS**

###**Advanced Cluster Profiling – Behavioral + Role Insights**

```python
from collections import Counter

def get_common_entries(df, cluster_label, column_name, top_n=3):
    cluster_data = df[df['kmeans_cluster'] == cluster_label]
    return Counter(cluster_data[column_name]).most_common(top_n)

cluster_profiles = {}

for cluster in sorted(df5['kmeans_cluster'].unique()):
    job_positions = get_common_entries(df5, cluster, 'job_position')
    companies = get_common_entries(df5, cluster, 'company_hash')

    cluster_profiles[cluster] = {
        'average_ctc': round(cluster_averages.loc[cluster, 'ctc_capped'], 2),
        'average_yoe': round(cluster_averages.loc[cluster, 'YOE_capped'], 2),
        'average_ctc_updates': round(cluster_averages.loc[cluster, 'no_of_ctc_update'], 2),
        'common_job_positions': job_positions,
        'common_companies': companies
    }

# Display Profiles
for cluster, profile in cluster_profiles.items():
    print("="*50)
    print(f"📌 Cluster {cluster} Profile")
    print("="*50)
    print(f"💰 Avg Compensation (CTC): {profile['average_ctc']}")
    print(f"📈 Avg Years of Experience: {profile['average_yoe']} years")
    print(f"🔄 Avg CTC Updates: {profile['average_ctc_updates']}")

    print("\n👨‍💼 Top Job Positions:")
    for job, count in profile['common_job_positions']:
        print(f"   • {job} ({count})")

    print("\n🏢 Top Companies:")
    for company, count in profile['common_companies']:
        print(f"   • {company} ({count})")
    print("\n")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score
import pandas as pd

# Copy original dataframe
df8 = df3.copy()

# Step 1: Perform K-Means
kmeans = KMeans(n_clusters=5, random_state=42)

df8['Cluster'] = kmeans.fit_predict(
    df8[['ctc_capped', 'YOE_capped',
         'designation_flag', 'class_flag', 'tier_flag']]
)

# Step 2: Create Manual Cluster Label
# Combine manual flags into a single label
df8['Manual_Cluster'] = (
    df8['designation_flag'].astype(str) +
    df8['class_flag'].astype(str) +
    df8['tier_flag'].astype(str)
)

# Convert to categorical integer codes
df8['Manual_Cluster'] = pd.Categorical(df8['Manual_Cluster']).codes

# Step 3: Evaluate Similarity
ari = adjusted_rand_score(df8['Manual_Cluster'], df8['Cluster'])
nmi = normalized_mutual_info_score(df8['Manual_Cluster'], df8['Cluster'])

print(f'Adjusted Rand Index (ARI): {ari:.4f}')
print(f'Normalized Mutual Information (NMI): {nmi:.4f}')
```
